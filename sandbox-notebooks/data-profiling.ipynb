```python
def count_customers_with_multiple_subscriptions(df):
    # Count unique subscription_id per customer_id and from_date
    customer_counts = df.groupby(['customer_id', 'from_date'])['subscription_id'].nunique()
    
    # Count how many customers have more than one subscription with the same from_date
    count = (customer_counts > 1).sum()
    
    return count
```

```python
count_customers_with_multiple_subscriptions(df)
```
